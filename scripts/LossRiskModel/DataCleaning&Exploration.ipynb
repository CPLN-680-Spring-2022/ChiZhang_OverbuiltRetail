```python
centerinfo = pd.read_excel("C:\\Users\\Chi\\Desktop\\shanghai\\exploration.xls")
centerinfo['Coordinates'] = gpd.points_from_xy(centerinfo['lon'], centerinfo['lat'])
centerinfo = gpd.GeoDataFrame(centerinfo, 
                            geometry="Coordinates", 
                            crs="EPSG:4326")
centerinfo.head()
```

```python
census = gpd.read_file("C:\\Users\\Chi\\Desktop\\shanghai\\GeoData\\census_1984.shp")
census = census.to_crs(epsg=4326)
census.head()
len(census)
```

```python
census.head()
```

```python
#Spatially join data sets
joined = gpd.sjoin(centerinfo, census, op='within', how='left')
joined.head()
```

```python
#center_g = pd.DataFrame(joined.drop(columns='geometry'))
center_g=joined.loc[:,["area","FID_1"]]
center_g['FID_1'] = center_g['FID_1'].astype('category')
accumulate =center_g.groupby('FID_1')['area'].agg(['sum','count']).rename(columns={'sum':'total_area'})
accumulate.head()
```

```python
total = census.merge(accumulate, on='FID_1',how="left")
total['total_area'] = total['total_area'].fillna(0)
total['count'] = total['count'].fillna(0)
total['POP_density'] = total['POP']/total['Shape_Area']/10000
total['Center_density'] = total['total_area']/total['Shape_Area']/10000
total.head()
```

```python
#Spatially join data sets
joined2 = gpd.sjoin(centerinfo, total, op='within', how='left')
joined2.head()
```

```python
color_map = {"Shopping Centers": "#F172A1"}
variables = [
    "area",
    "grading",
    "year",
    "AvgConsump",
    "POP",
]

# Set the seaborn style
sns.set_context("talk", font_scale=0.75)

# make the pair plot
sns.pairplot(
    joined[variables].dropna(),
    plot_kws=dict(alpha=0.2, edgecolor="none"),
)
```

```python
#correlation validation
cols = [
    "area",
    "grading",
    "year",
    "POP",
    "AvgConsump",
    "POP_density",
    "Center_density",
    "count"
]

plt.figure(figsize=(15, 10))
sns.heatmap(joined2[cols].corr(), cmap='coolwarm', annot=True, vmin=-1, vmax=1);
```

```python
#plot hist
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(joined["grading"],30, density=True, facecolor='#C00000', alpha=0.75)

# Format the axes
ax.set_xlabel("Grading")
ax.set_ylabel("Number of gradings")
ax.set_axisbelow(True)
ax.grid(True, axis="y",alpha=0.5,linestyle="dotted")
ax.set_yticklabels([f"{yval:,.0f}" for yval in ax.get_yticks()] )
ax.set_title("Shopping center customer gradings")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add the NDVI = 0 threshold
ax.axvline(x=4.3, c='k', linewidth=2)
ax.text(4.3, 1, "Average grades", ha='left', fontsize=14);
```
