```python
from urllib.request import urlopen,quote
import requests,csv
import pandas as pd
import json
```

```python
def getlnglat(adress):
    url = 'http://api.map.baidu.com/geocoder/v2/?address='
    output = 'json'
    ak = 'ikUx7vHBNmVThOGrrvsMsttaP6XByDCE'
    add = quote(adress)#使用quote进行编码 为了防止中文乱码
    url2 = url + add + '&output=' + output + '&ak=' + ak
    req = urlopen(url2)
    res = req.read().decode()
    temp = json.loads(res)
    return temp
```

```python
import requests,csv
import pandas as pd
import json
import requests
import json
import pprint
ak = 'fRIm8SVgC3ZcNmK7nSXz2Ha9GZInPpQm'#ak需要去百度地图申请
 
def getlnglat(address):
    url = 'https://api.map.baidu.com/geocoding/v3/?address={}&output=json&ak={}'.format(address,ak)
    res = requests.get(url)
    json_data = json.loads(res.content)
    return json_data
```
